```octave
fc=10e3;
wc=fc*2*pi;
sf = 40e3; 
sw=sf*2*pi;
data=[[1;zeros(sf-1,1)],sinetone(2e3,sf,1,1),sinetone(10e3,sf,1,1),sinetone(25e3,sf,1,1)];
[b,a]=butter ( 1, wc / sw );
filtered = filter(b,a,data);

clf
subplot ( columns ( filtered ), 1, 1)
plot(filtered(:,1),";Impulse response;")
subplot ( columns ( filtered ), 1, 2 )
plot(filtered(:,2),";2kHz response;")
subplot ( columns ( filtered ), 1, 3 )
plot(filtered(:,3),";10kHz response;")
subplot ( columns ( filtered ), 1, 4 )
plot(filtered(:,4),";25kHz response;")
```
